```python
from augments import (
    Classification,
    KeywordsExtractor,
    ParagraphGenerator,
    ParagraphTrimmer,
    Reasoning,
    Summarizer,
    WikipediaCrawler,
)
```

NOTE: 전체 실험 기록은 [여기](https://gamchan.notion.site/146815b39d39807884f1f785c2829da6?v=086629b8bb6f49a5a1b54a0ec44d6630&pvs=4)에서 참고하실 수 있습니다


외부 지식이 필요한지 여부를 판단하기 앞서, 문제와 답변을 주고 풀이 과정을 추론하게 한다. [관련 실험 링크](https://gamchan.notion.site/reasoning-1-149815b39d39802b8c58d85ada0525b7?pvs=4)


```python
reasoning = Reasoning()
sampled = 5  # 5개 샘플로 실험
reasoning.sample_data(n=sampled)
reasoning.process()
reasoning.data  # 풀이 과정 추론 확인
```

문제 풀이 과정과 문제를 통해 외부 지식이 필요한지 여부를 확인한다. [관련 실험 링크](https://gamchan.notion.site/Reasoning-2-14a815b39d3980649269ed36d2cf1b25?pvs=4)

- 내적 추론: 외부 지식이 필요 없는 문제
- 외적 추론: 외부 지식이 필요한 문제


```python
classification = Classification(data=reasoning.data)
classification.process()
classification.data  # 분류 확인
```

분류한 문제 중 외부 지식이 필요한 문제에서 위키피디아 문서 검색 키워드를 추출한다. [관련 실험 링크](https://gamchan.notion.site/reasoning-2-149815b39d39806e8d84c19f6b1d1ef1?pvs=4)


```python
keyword_extraction = KeywordsExtractor(data=classification.data)
keyword_extraction.process()
keyword_extraction.data  # 추출된 키워드 확인
```

추출된 검색 키워드로 해당 위키피디아 문서를 크롤링한다. (단순 크롤링이므로 실험 기록 없음)


```python
wikipedia_crawler = WikipediaCrawler(data=keyword_extraction.data)
wikipedia_crawler.crawl()
wikipedia_crawler.data  # 크롤링된 문서 확인
```

키워드 별 크롤링된 문서를 추출한 키워드와 같이 요약한다. [관련 실험 링크](https://gamchan.notion.site/5-14a815b39d39806a9dbce00233fe3755?pvs=4)


```python
summarizer = Summarizer(data=wikipedia_crawler.data)
summarizer.process()
summarizer.data  # 요약된 키워드 별 문서 확인
```

요약문을 하나의 지문으로 취합한다. [관련 실험 링크](https://gamchan.notion.site/4-149815b39d3980e88d73eb48e438a28a?pvs=4)(NOTE: 첨부 링크에서는 지문 취합과 지문 다듬기 실험을 동시에 진행했습니다.)


```python
paragraph_generator = ParagraphGenerator(data=summarizer.data)
paragraph_generator.process()
paragraph_generator.data  # 취합한 지문 확인
```

취합한 지문을 수능 비문학 국어 형식으로 다듬는다.


```python
paragraph_trimmer = ParagraphTrimmer(data=paragraph_generator.data)
paragraph_trimmer.process()
```
